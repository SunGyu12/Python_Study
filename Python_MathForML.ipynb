# 미분

f(x) = x**2            x에 대한 식(1)
dx_f(x) = 2x           ..식(1)에 대해서 x에 대한 1차 미분

f(x,w) = x**2+w**2    x에 대한 식(2)
dx_f(x,w) = 2x         ..식(2)에 대해서 x에 대한 1차 미분
dx_f(x,w) = 2w         ..식(2)에 대해서 w에 대한 1차 미분

```python
def f(w0, w1):
    return w0**2 + 2*w0*w1 + 3  # x^2 + 2xy + 3
def df_dw0(w0, w1):
    return 2*w0 + 2*w1          # 2x + 2y
def df_dw1(w0, w1):
    return 2*w0                 # 2x
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def f(w0, w1):
    return w0**2 + 2*w0*w1 + 3

w_range = 2
dw=0.25
w0=np.arange(-w_range, w_range+dw, dw)
w1=np.arange(-w_range, w_range+dw, dw)
ww0, ww1 = np.meshgrid(w0, w1)
y = f(ww0, ww1)

fig = plt.figure()
ax = plt.subplot(1,1,1, projection='3d')
ax.plot_surface(ww0, ww1, y, rstride = 1, cstride = 1, alpha = 0.3,
                color='blue', edgecolor='black')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(w0, w1):
    return w0**2 + 2*w0*w1 + 3  # x^2 + 2xy + 3

def df_dw0(w0, w1):
    return 2*w0 + 2*w1          # 2x + 2y

def df_dw1(w0, w1):
    return 2*w0                 # 2x

w_range = 10
dw = 1
w0 = np.arange(-w_range, w_range+dw, dw)
w1 = np.ones(len(w0))
w1 *= -1

titles = ['$w_1=-1$','$w_1=1$','$w_0=-1$','$w_0=1$']
funcs = [f, f, lambda w0, w1: f(w1, w0), lambda w0, w1: f(w1, w0)]
derivatives = [df_dw0, df_dw0, df_dw1, df_dw1]

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.ravel()

for i in range(4):
    axs[i].plot(w0, funcs[i](w0, w1), 'black', linewidth=3, label='$f(w_0, w_1)$')
    axs[i].plot(w0, derivatives[i](w0, w1), 'cornflowerblue', linewidth=3, label='$\\frac{\\partial f}{\\partial w_'+ str(i % 2) +'}$')
    axs[i].set_title(titles[i])
    axs[i].set_xlabel('$w_' + str(i % 2) +'$')
    axs[i].set_ylabel('$y$')
    axs[i].set_ylim(-10, 10)
    axs[i].set_xlim(-5, 5)
    axs[i].grid(True)
    axs[i].legend(loc='lower right')

plt.show()
```

```python
#exp graph
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4,4,100)

y = 2**x
y2 = 3**x
y3 = 0.5**x

plt.figure(figsize=(3,3))
plt.plot(x,y,'black', linewidth=3, label='$y=2^x$')
plt.plot(x,y2,'cornflowerblue', linewidth=3, label='$y=3^x$')
plt.plot(x,y3,'gray', linewidth=3, label='$y=0.5^x$')
plt.ylim(-2,6)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```
