```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline                                                  #모듈 추가
x = np.linspace(-4,4, 100)
y1 = 2**x
y2 = 3**x
y3 = 0.5**x                                                        #데이터 생성

print('x = {}'.format(x))
print('y1 = {}'.format(y1))
print('y2 = {}'.format(y2))
print('y3 = {}'.format(y3))                                        #데이터 log확인

plt.figure(figsize=(5,5))
plt.plot(x,y1, 'black', linewidth = 3, label = '$y=2^x$')
plt.plot(x,y2, 'cornflowerblue', linewidth = 3, label = '$y=3^x$')
plt.plot(x,y3, 'gray', linewidth = 3, label = '$y=0.5^x$')
plt.ylim(-2,6)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='lower right')
#plt.show()
plt.savefig('graph.png')                                           #그래프 출력
```

# 3차 함수 f(x) = (x-2)*x*(x+2)

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    y = (x-2)*x*(x+2)
    return y

print('x = 1', f(1))
print('x = 2', f(2))
print('x = 3', f(3))

x = np.array([1,2,3])     #list -> np.ndarray
x = np.arange(-3,3.5,0.5) #np.arange -> 범위 지정
x = np.linspace(-3,3,100)  #start, finishm, samples
print('x [] = {}, type={}'.format(x, type(x)))
print(f(x))
plt.plot(x,f(x))

plt.figure(figsize=(5,5))
plt.plot(x,f(x), 'cornflowerblue', linewidth = 3, label = '$y=3^x$')
plt.ylim(-30, 30)
plt.ylim(-2,6)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
plt.savefig('graph.png')                            
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    y = (x-2)*x*(x+2)
    return y

def f2(x,w):
    y = (x-w)*x*(x+2)
    return y

x = np.linspace(-3,3,50)
print('x [] = {}, type={}'.format(x, type(x)))
print(f(x))

plt.figure(figsize=(5,5))
plt.plot(x,f(x), 'cornflowerblue', linewidth = 3, label = '$y = (x-2)*x*(x+2)$')

plt.plot(x,f2(x,w=0), 'black', linewidth = 3, label = '$w=0$')
plt.plot(x,f2(x,w=4), 'gray', linewidth = 3, label = '$y = (x-4)*x*(x+2)$')

plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.ylim(-30, 30)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
plt.savefig('graph.png')                            
```

```python
def f3(x0, x1):
    r = 2*x0**2 + x1**2
    ans = r*np.exp(-r)
    return ans

#x0, x1에 따라서 각각 f3함수를 계산
xn = 9
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((len(x0),len(x1)))
print('x0 = {}'.format(x0))
print('x1 = {}'.format(x1))
print('y = {}'.format(y))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])

y = np.round(y,1)
print('y = {}'.format(y))
plt.figure(figsize = (3.5, 3))
plt.gray()
#plt.jet()
#plt.pink()
#plt.bone()

#plt.pcolor(y)
plt.contour(y)

plt.colorbar()
#plt.show()
plt.savefig('gauss.png')
```

# Quiz 구구단 3, 5, 7, 9단 Plot(1개), SubPlot(4분할)로 표현

```python
import matplotlib.pyplot as plt

def GuGu(dan):
    x = np.linspace(1,9, 9)
    y = dan*x
    return y

x = np.linspace(1,9, 9)
plt.figure(figsize = (5,5))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

subPlotIndex = 1
for i in range(3,10,2):
    plt.subplot(2,2,subPlotIndex)#plot 묘사의 위치 지정
    titleName = '{}Dan'.format(i)
    plt.title(titleName)
    plt.plot(x, GuGu(i),'k')
    plt.ylim(0,91)
    plt.grid(True)
    subPlotIndex+=1
plt.show()

plt.figure(figsize=(5,5))

plt.plot(x,GuGu(3), 'cornflowerblue', linewidth = 3, label = '$y=3*x$')
plt.plot(x,GuGu(5), 'black', linewidth = 3, label = '$y=5*x$')
plt.plot(x,GuGu(7), 'gray', linewidth = 3, label = '$y=7*x$')
plt.plot(x,GuGu(9), 'yellow', linewidth = 3, label = '$y=9*x$')




plt.title('$GuGu(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.ylim(1,81)
plt.xlim(1,9)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
